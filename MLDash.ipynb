```python
from Regression import Multi_nom_reg
from DecisionTree import DecisionTreeClass
from FNN import CNNClass
```

```python
from dash import Dash, dcc, html, Input, Output, callback
```

```python
def DecisionTree_plot():
  DecisionTreeModel = DecisionTreeClass()
  DecisionTreeModel.loadDataset('NFLPlaybyPlay2015.csv')
  DecisionTreeModel.cleanData()
  DecisionTreeModel.trainTestDataset()
  DecisionTreeModel.createDecisionTreeModel()
  DecisionTreeModel.predict_proba()
  return DecisionTreeModel.ROC_curve()

def MNR_plot():
  MNR = Multi_nom_reg()
  MNR.loadDataset('NFLPlaybyPlay2015.csv')
  MNR.cleanData()
  MNR.trainTestDataset()
  MNR.RegModel(max_iter=800)
  MNR.predict_proba()
  return MNR.ROC_curve()

def CNN_plot():
    cnnModel = CNNClass()
    cnnModel.loadDataset('NFLPlaybyPlay2015.csv')
    cnnModel.cleanData()
    cnnModel.trainTestDataset()
    cnnModel.createCnnModel(12, 4)
    cnnModel.trainModel(numEpochs=10)
    return cnnModel.performanceMeasure()
    
```

```python
app = Dash(__name__)
```

```python
app.layout = html.Div([

    html.H1('NFL Model Comparisons',
            style = {'color':'white', 'fontsSze': '40px'}),

    dcc.Dropdown(
        id="machine_learning_model_dropdown",
        options=['Decision Tree', 'Regression', 'CNN'],
        value='Regression',
        style={'color': 'blue'}
    ),

    html.Div(style={'padding': '20px'}),

   dcc.Graph(id = 'output-container-graph')
])
```

```python
@callback(
    Output('output-container-graph', 'figure'),
    Input('machine_learning_model_dropdown', 'value')
)

def update_output(value):
    match value:
        case 'CNN':
            print('calling on CNN')
            return CNN_plot()
        case 'Decision Tree':
            return DecisionTree_plot()
        case 'Regression':
            return MNR_plot()
        case _:
            return {}
```

```python
if __name__ == '__main__':
    app.run(debug=True)
```
